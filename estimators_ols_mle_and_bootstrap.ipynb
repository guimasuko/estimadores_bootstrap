## Exercício 1
Considere os três modelos a seguir e uma amostra aleatória de tamanho $n\left\{y_{i}^{(s)}, x_{i}\right\}_{i=1}^{n}$,

1. Modelo linear com uma mistura de distribuições normais.
$$
\begin{aligned}
	&y_{i}^{(1)}=\beta_{0}+\beta_{1} x_{i}+e_{i} \\
	&e_{i} \sim \mathbb{D}_{\epsilon}=p \mathcal{N}\left(\mu_{1}, \sigma_{1}^{2}\right)+(1-p) \mathcal{N}\left(\mu_{2}, \sigma_{2}^{2}\right)
\end{aligned}
$$
onde $\mathcal{N}\left(\mu, \sigma^{2}\right)$ é a de uma distribuição Normal com média $\mu$ e variância $\sigma^{2}$, e $p=1 / 2$.

2. Modelo linear com inovações Gamma.
$$
\begin{aligned}
	&y_{i}^{(2)}=\beta_{0}+\beta_{1} x_{i}+u_{i} \\
	&u_{i} \sim \Gamma(a, b)
\end{aligned}
$$

3. Modelo linear com uma distribuição Cauchy padrão.
$$
\begin{aligned}
	&y_{i}^{(3)}=\beta_{0}+\beta_{1} x_{i}+v_{i} \\
	&v_{i} \sim t_{(1)}(0,1)
\end{aligned}
$$
Para cada modelo, escreva a função de verossimilhança amostral, os estimadores de máxima verossimilhança e suas distribuições assintóticas.


### Resposta

Considerando uma amostra aleatória de tamanho $n$ denotada por $\mathbf{z}^{(s)}=\left\{y_{i}^{(s)}, x_{i}\right\}_{i=1}^{n}$. 

Para as derivações a seguir, supõe-se que a função densidade de probabilidade $f(x)$ não depende de $\theta^{(i)}$, onde $\theta^{(i)}, i=\{1,2,3\}$, é o vetor de parâmetros a ser definido em cada item abaixo. 

Para obter a densidade condicional em $X$ dos $y^{(i)}$ a partir da densidade do erro, precisamos fazer uma transformação. Supondo uma função $g$ bem definida, pois $g(y|x)$ é linear e estritamente crescente em y, onde 

\begin{equation*}
\begin{split}
	y | x=g(e)=e+\beta X\\
	\Leftrightarrow g^{-1}(y \mid x)=y-\beta X
\end{split}
\end{equation*}
e, assim, 

\begin{equation*}
\begin{split}
	f(y \mid X)=f_{e}\left(g^{-1}(y)\right) \frac{\partial g^{-1}(y)}{\partial y}=f_{e}(y-\beta X)
\end{split}
\end{equation*}

Ao longo da lista, $\beta=\left(\begin{array}{ll}\beta_{0} & \beta_{1}\end{array}\right)'$ é o vetor dos parâmetros beta, e x é o vetor amostral cujo iésimo elemnto é o x realizado na $n$ observação amostral, e $X=\left(\begin{array}{ll}1_{n x 1} & x\end{array}\right)$ é o vetor de regressores, composto por um vetor de constantes e o vetor $x$.

### 1.1 Modelo linear com uma mistura de distribuições normais
\begin{equation*}
\begin{aligned}
	&y_{i}^{(1)}=\beta_{0}+\beta_{1} x_{i}+e_{i} \\
	&e_{i} \sim \mathbb{D}_{e}=p \mathcal{N}\left(\mu_{1}, \sigma_{1}^{2}\right)+(1-p) \mathcal{N}\left(\mu_{2}, \sigma_{2}^{2}\right)
\end{aligned}
\end{equation*}
onde $\mathcal{N}\left(\mu, \sigma^{2}\right)$ é uma distribuição Normal com média $\mu$ e variância $\sigma^{2}$ e $p=\frac{1}{2}$.

Seja $\theta^{(1)}=\left(\beta_{0}, \beta_{1}, \mu, \sigma\right)$ o vetor composto pelos parâmetros. Note que $Y \mid\left(X, \theta^{(1)}\right) \sim \mathcal{N}\left(\beta_{0}+\beta_{1} X+\mu_{1}, \sigma_{1}^{2}\right)$ com probabilidade $\frac{1}{2}$ e $Y \mid\left(X, \theta^{(1)}\right) \sim \mathcal{N}\left(\beta_{0}+\beta_{1} X+\mu_{2}, \sigma_{2}^{2}\right)$ com probabilidade $\frac{1}{2}$. Assim, temos a seguinte função verossimilhança amostral:

\begin{equation*}
\begin{aligned}
	&L\left(\theta^{(1)}\right)=f\left(\mathbf{z}^{(1)} ; \theta^{(1)}\right)=\prod_{i=1}^{n} f\left(\mathbf{z}_{i}^{(1)} ; \theta^{(1)}\right)=\prod_{i=1}^{n} f\left(y_{i}^{(1)} \mid x_{i}, \theta^{(1)}\right) f\left(x_{i}\right)= \\
	&\prod_{i=1}^{n}\left\{\frac{1}{2} \frac{1}{\sigma_{1} \sqrt{2 \pi}} \exp \left(\frac{-\left(y_{i}^{(1)}-\left(\beta_{1} x_{i}+\beta_{0}+\mu_{1}\right)\right)^{2}}{2 \sigma_{1}^{2}}\right) f\left(x_{i}\right)+\frac{1}{2} \frac{1}{\sigma_{2} \sqrt{2 \pi}} \exp \left(\frac{-\left(y_{i}^{(1)}-\left(\beta_{1} x_{i}+\beta_{0}+\mu_{2}\right)^{2}\right.}{2 \sigma_{2}^{2}}\right) f\left(x_{i}\right)\right\}
\end{aligned}
\end{equation*}
Transformando para obter a função log verossimilhança:




Os estimadores de máxima verossimilhança são tais que:

\begin{equation*}
\hat{\theta^{(1)}}=\operatorname{argmax}_{\theta^{(1)} \in \Theta^{(1)}} f\left(\mathbf{z}^{(1)} ; \theta^{(1)}\right)=\operatorname{argmax}_{\theta^{(1)} \in \Theta^{(1)}} \log L\left(\theta^{(1)}\right)
\end{equation*}
pois a função log é uma transformação monotônica.


### 1.2 Modelo linear com inovações Gamma
\begin{equation*}
\begin{aligned}
	&y_{i}^{(2)}=\beta_{0}+\beta_{1} x_{i}+u_{i} \\
	&u_{i} \sim \Gamma(a, b)
\end{aligned}
\end{equation*}

Seja $\theta^{(2)}=\left(\beta_{0}, \beta_{1}, a, b\right)$ o vetor de parâmetros.
Primeiro, note que a densidade do erro é igual a zero para $e^{(2)}<0$, o que transformando nos dá, $y_{i}^{(2)}-\beta X<0$.
Assim, a função de verossimilhança amostral é:
\begin{equation*}
\begin{aligned}
	L\left(\theta^{(2)}\right)&=f\left(\mathbf{z}^{(2)} ; \theta^{(2)}\right)\\
	&=\prod_{i=1}^{n} f\left(y_{i}^{(2)} \mid x_{i}, \theta^{(2)}\right) f\left(x_{i}\right) \\
	&=\prod_{i=1}^{n}\left\{\frac{\left(y_{i}^{(2)}-\beta_{0}-\beta_{1} x_{i}\right)^{a-1}}{\Gamma(a) b^{a}} \exp \left(\frac{-\left(y_{i}^{(2)}-\beta_{0}-\beta_{1} x_{i}\right)}{b}\right)\right\}, y_{i}^{(2)}-\beta X \geq 0
\end{aligned}
\end{equation*}
Aplicando a transformação do log, obtemos, definido só quando $y_{i}^{(2)}-\beta X \geq 0$ :

\begin{equation*}
\log L\left(\theta^{(2)}\right)=-n \log \left(\Gamma(a) b^{a}\right)+(a-1) \sum_{i=1}^{n} \log \left(y_{i}^{(2)}-\beta_{0}-\beta_{1} x_{i}\right)-\frac{1}{b} \sum_{i=1}^{n}\left(y_{i}^{(2)}-\beta_{0}-\beta_{1} x_{i}\right)
\end{equation*}

Os estimadores de máxima verossimilhança são definidos como:

\begin{equation*}
\hat{\theta^{(2)}}=\operatorname{argmax}_{\theta^{(2)} \in \Theta^{(2)}} f\left(\mathbf{z}^{(2)} ; \theta^{(2)}\right)=\operatorname{argmax}_{\theta^{(2)} \in \Theta^{(2)}} \log L\left(\theta^{(2)}\right)
\end{equation*}

### 1.3 Modelo linear com uma distribuição Cauchy padrão
\begin{equation*}
\begin{aligned}
	&y_{i}^{(3)}=\beta_{0}+\beta_{1} x_{i}+v_{i} \\
	&v_{i} \sim t_{(1)}(0,1)
\end{aligned}
\end{equation*}

Seja $\theta^{(3)}=\left(\beta_{01} \beta_{1}, r\right)$. Função verossimilhança amostral:

\begin{equation*}
\begin{aligned}
	L\left(\theta^{(3)} \mid r=1\right)&=f\left(\mathbf{z}^{(3)} ; \theta^{(3)} \mid r=1\right)\\
	&=\prod_{i=1}^{n} f\left(y_{i}^{(3)} \mid x_{i}, \theta^{(3)}, r=1\right) f\left(x_{i}\right) \\
	&=\prod_{i=1}^{n}\left\{\frac{1}{\pi\left(1+\left(y_{i}^{(3)}-\beta_{0}-\beta_{1} x_{i}\right)^{2}\right)} f\left(x_{i}\right)\right\}
\end{aligned}
\end{equation*}

Aplicando o log, temos: 

\begin{equation*}
\log L\left(\theta^{(3)} \mid r=1\right)=\sum_{i=1}^{n}\left\{\log \frac{1}{\pi\left(1+\left(y_{i}^{(3)}-\beta_{0}-\beta_{1} x_{i}\right)^{2}\right)}\right\}+\sum_{i=1}^{n} f\left(x_{i}\right)
\end{equation*}

Os estimadores de máxima verossimilhança são definidos como:

\begin{equation*}
\hat{\theta^{(3)}}=\operatorname{argmax}_{\theta^{(3)} \in \Theta^{(3)}} f\left(\mathbf{z}^{(3)} ; \theta^{(3)} \mid r=1\right)=\operatorname{argmax}_{\theta^{(3)} \in \Theta^{(3)}} \log L\left(\theta^{(3)} \mid r=1\right)
\end{equation*}

### 1.4 Distribuições Assintóticas

Pelo Teorema $6.3$ dado em sala de aula - teorema 14.1 em Greene (2012) - sabemos que, sob certas condições de regularidade, se os estimadores de máxima verossimilhança estão identificados, $\sqrt{n}\left(\hat{\theta}_{mv}-\theta_{0}\right) \sim \mathcal{N}\left(0,\left(\mathcal{F}\left(\theta_{0}\right)\right)^{-1}\right)$, onde $\mathcal{F}\left(\theta_{0}\right)=-\mathbb{E}\left[\frac{\partial^{2}}{\partial \theta_{0} \partial \theta_{0}^{\prime}} \ln L\left(\theta_{0}\right)\right]$ (matriz de informação de Fisher), e $\theta_{0}$ é o valor populacional do vetor de parâmetros. As condições são as seguintes:

* R1. As três primeiras derivadas de $\ln f\left(z_{i} \mid \theta\right)$ em $\theta$ são contínuas e finitas para quase todo $z_{i}$ e todo $\theta$.
*  R2. Condições necessárias para obter esperança da primeira e segunda derivadas de $\ln f\left(\mathbf{z}_{i} \mid \theta\right)$ são válidas.
* R3. Para todos os valores de $\theta,\left|\frac{\partial^{3} \mid \ln f\left(z_{i} \mid \theta\right)}{\partial \theta_{j} \partial \theta_{k} \partial \theta_{\ell}}\right|$ é menos que uma função que tem esperança finita.
\end{itemize}

Uma condição suficiente para que valha a segunda condição de regularidade é que o suporte da distribuição não dependa de $\theta$, o que é o caso no modelo 1 e no modelo 3 , mas não no modelo 2 , o que não significa que R2 não valha, apenas que seria necessário realizar uma investigação maior. As duas outras exigem derivações além do escopo dessa lista, embora o fato das densidades vistas acima serem funções compostas de funções que atendem as condições de regularidade pareça indicar sua validade.

Em probabilidade e estatística, uma distribuição de mistura é a distribuição de probabilidade de uma variável aleatória que é derivada de uma coleção de outras variáveis ​​aleatórias como segue: primeiro, uma variável aleatória é selecionada ao acaso da coleção de acordo com determinadas probabilidades de seleção, e então o valor da variável aleatória selecionada é realizado.

```python
# primeiro definimos um vetor aleatórios de números 0 ou 1
escolha = np.random.choice([1,0], size = 500, replace = True, p = [0.5, 0.5])
# o termo de erro do primeiro modelo é uma variável aletória que assume o valor 
e = escolha*e1 + (np.ones(500) - escolha)*e2
```

```python
# Modelo 1
y1 = b0 + b1 * x + e
```

```python
# Plots
plt.figure(figsize=(10, 8), dpi=80)
sns.histplot((e, y1), kde=True, binwidth=1, stat='density')
plt.legend(title='Dados', loc='upper right', labels=['y1', 'e'])
```

A variável $e$ é composta pela média de duas normais com médias $\mu_1 = -2$ e $\mu_2 = 2$, e variâncias $\sigma_1^2 = 0.5$ e $\sigma_2^2 = 25$, respectivamente, ela parece ter média em 0, e sua calda à direita parece ser mais longa do que à esquerda. Isso se deve ao fato da normal com a média maior ter uma variância maior do que a normal com média menor. Os valores com mais probabilidade ficam mais próximo da média da primeira normal, isso acontece porque a variância menor relativa dessa amostra representa probabilidades maiores de se obter valores próximos à média, isso em relação a segunda normal.

```python
# Algumas estatísticas descritivas do Modelo 1
stats_1 = pd.DataFrame([[np.mean(x), statistics.median(x), np.var(x), min(x), max(x)],
                        [np.mean(y1), statistics.median(y1), np.var(y1), min(y1), max(y1)],
                        [np.mean(e), statistics.median(e), np.var(e), min(e), max(e)]],
                        columns=['Média','Mediana','Variância','Mínimo','Máximo'], 
                        index= ['x','y1','e'])
stats_1                    
```

Para ver se esses valores são coerentes, precisamos analisar os valores populacionais dessas variáveis.
\begin{equation*}
\begin{aligned}
	&\mathbb{E}\left(e\right)=p \mathbb{E}\left(\mathcal{N}_{(1,1)}\right)+(1-p) \mathbb{E}\left(\mathcal{N}_{(1,2)}\right)=0 \\
	&\mathbb{E}\left(y^{(1)}\right)=1.5+3 \mathbb{E}(x)+\mathbb{E}\left(e^{(1)}\right)=16.5 \\
\end{aligned}
\end{equation*}

Vou dar uma atenção a mais aos valores populacionais das variâncias, isso porque é a primeira vez que estamos lidando com uma mistura de variáveis aleatórias.

Primeiro, vamos relembrar a simples formula da variância

\begin{equation*}
\begin{aligned}
    &\operatorname{Var}\left(e\right) = \mathbb{E}\left(e^2\right) - \mathbb{E}\left(e\right)^2
\end{aligned}
\end{equation*}

Vamos atrás do primeiro e segundo momento de $e$, respectivamente $\mathbb{E}\left(e\right)$ e $\mathbb{E}\left(e^2\right)$,

\begin{equation*}
\begin{aligned}
    &\mathbb{E}\left(e\right) = p\mathbb{E}\left(e_1\right) + (1-p)\mathbb{E}\left(e_2\right) = \frac{1}{2}\cdot(-2) + \frac{1}{2}\cdot(2) = 0\\
\end{aligned}
\end{equation*}

Para o cálculo do segundo momento de $e$, há a necessidade de antes, calcular os segundos momentos de $e_1$ e $e_2$, $\mathbb{E}\left(e_1^2\right)$ e $\mathbb{E}\left(e^2_2\right)$

\begin{equation*}
\begin{aligned}
    &\mathbb{E}\left(e_1^2\right) = \operatorname{Var}\left(e_1\right) + \mathbb{E}\left(e_1\right)^2 = 0.5 + (-2)^2 = 4.5 \\
    &\mathbb{E}\left(e_2^2\right) = \operatorname{Var}\left(e_2\right) + \mathbb{E}\left(e_2\right)^2 = 25 + 2^2 = 29
\end{aligned}
\end{equation*}

e, portanto,

\begin{equation*}
\begin{aligned}
    &\mathbb{E}\left(e^2\right) = p\mathbb{E}\left(e_1^2\right) + (1-p)\mathbb{E}\left(e_2^2\right) = \frac{1}{2}\cdot(4.5) + \frac{1}{2}\cdot(29) = 16.75
\end{aligned}
\end{equation*}

Voltando a fórmula da variância

\begin{equation*}
\begin{aligned}
    &\operatorname{Var}\left(e\right) = \mathbb{E}\left(e^2\right) - \mathbb{E}\left(e\right)^2 = 16.75 - 0^2 = 16.75
\end{aligned}
\end{equation*}

Obtendo a variância de $y^(1)$
\begin{equation*}
\begin{aligned}
	&\operatorname{Var}\left(y^{(1)}\right)=\operatorname{Var}\left(\beta_{0}+\beta_{1} x+e\right) =\beta_{1}^2 \sigma_{x}^{2}+\operatorname{var}\left(e\right)=3^2\cdot(1) + 16.75 = 25.75
\end{aligned}
\end{equation*}


Os valores amostrais parecem bem próximos dos populacionais.

A distribuição de Cauchy é particular por se tratar de uma distribuição que não tem nem esperança, nem variância bem definidas. Embora os valores sejam similares e a distribuição de $y^{(3)}\mid x$ seja uma Cauchy, a distribuição de $y^{(3)}$ incondicional deve ser outra distribuição não podendo fazer quaisquer afirmações adicionais.

Sua distribuição aparece deslocada à direita da distribuição do erro porque tanto $\beta_0$ quanto $\beta_1$ tomam são valores positivos, enquanto a variável $x$ segue uma distribuição normal com média $\mu_x = 5$ e variância $\sigma^2_x = 1$, i.e. baixa probabilidade de receber valores negativos.

* A estimação de parâmetros por meio do método de Máximo Verossimilhaça depende totalmente da distribuição dos dados, uma vez que é utilizado a função densidade para chegar na função likelihood. Dessa forma, a melhor alternativa é conhecer bem qual distribuição de probabilidade os dados seguem. Quando não temos informação sobre a distribuição dos dados, o que nos resta é fazer suposições sobre estes afim de utilizar a distribuição Gaussiana. 

* Se os erros condicionais a $x$ seguem uma normal, os estimadores de $\hat{\beta}$ coincidem com os estimadores de OLS. Para mostrar isso, veja que, se $y \mid x \sim \mathcal{N}\left(\beta_{0}+\beta_{1} x\right)$, a função verossimilhança amostral é

\begin{equation*}
\begin{aligned}
\log L(\theta)&=\sum_{i=1}^{n}\log \left[f\left(y_{i} \mid x_{i}, \theta\right)+f\left(x_{i}\right)\right]\\
&=-\frac{n}{2} \log \left(2 \pi \sigma^{2}\right)+\sum_{i=1}^{n}\left[-\frac{1}{2 \sigma^{2}}\left(y_{i}-\beta X_{i}\right)^{2}\right]+\sum_{i=1}^{n} \log f\left(x_{i}\right)
\end{aligned}
\end{equation*}

Note que maximizar a expressão acima com respeito a $\beta$ equivale a maximizar $\sum_{i=1}^{n}\left[-\frac{1}{2 \sigma^{2}}\left(y_{i}-\beta X_{i}\right)^{2}\right]$, ou seja, minimizar $\sum_{i=1}^{n}\left[\frac{1}{2 \sigma^{2}}\left(y_{i}-\beta X_{i}\right)^{2}\right]$, cujo vetor minimizador é o mesmo que minimiza $\sum_{i=1}^{n}\left(y_{i}-\beta X_{i}\right)^{2}$, isto é, a soma de quadrados minimizada por OLS. Além disso, não conhecemos $\sigma^{2}$, de forma que precisamos estimá-lo também.
Da CPO:

\begin{equation*}
\begin{aligned}
\frac{\partial}{\partial \sigma^{2}} \log L=\frac{-n}{2 \sigma^{2}}+\frac{1}{2\left(\sigma^{2}\right)^{2}} \sum_{i=1}^{n}\left(y_{i}-\widehat{\beta} X_{i}\right)^{2}=0\\ \Leftrightarrow \hat{\sigma}_{m v}^{2}=n^{-1} \sum_{i=1}^{n}\left(y_{i}-\widehat{\beta} X_{i}\right)^{2}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
	\widehat{Q}_{x x} \stackrel{P}{\longrightarrow} Q_{x x}\\
	\widehat{Q}_{x y} \stackrel{P}{\longrightarrow} Q_{x y}\\
	\widehat{Q}_{x x}^{-1} \stackrel{P}{\longrightarrow} Q_{x x}^{-1}\\ \widehat{\mathrm{Q}}_{\mathrm{xe}} \stackrel{P}{\longrightarrow} \mathrm{Q}_{\mathrm{xe}}=0 
\end{aligned}
\end{equation*}

e assim $\widehat{\beta} \longrightarrow \beta$ conforme $n \rightarrow \infty .$

De maneira geral, vemos que o estimador de $\beta_1$ se aproxima bem do valore verdadeiro, enquanto as estimativas para o valor de $\beta_0$ estão mais distântes, com exceção ao primeiro modelo. Analisando o primeiro modelo e fazendo a ressalva que a construção da estimação OLS faz uso da matriz $X$, isto é, faz uso de um vetor de constantes iguais à 1, vemos que as estimativas são próximos dos parâmetros verdadeiros, isso acontece pelo fato do valor esperado do erro da distribuição do primeiro modelo ser zero, coerente com o uso do intercepto. No segundo modelo, temos que a esperança do erro da distribuição não é zero, superestimando a estimativa do intercepto. Por fim, no terceiro modelo, a estimativa do intercepto também é distante do valor verdadeiro e, como a distribuição do erro não tem média definida, é difícil dizer os motivos pelos quais há tal imprecisão.

* Para inferência sobre os parâmetros, podemos usar o teste t, com estatistica de teste $|T|=\left|\frac{\hat{\beta}-\beta_{0}}{s(\hat{\beta})}\right|$, onde $s(\hat{\beta})$ é o erro padrão de $\widehat{\beta}$, cuja distribuição assintótica é normal pode ser usado em amostras finitas se os erros fossem gaussianos, ou se a amostra é grande o suficiente para considerarmos razoável aproximar a distribuição do estimador pela sua distribuição assintótica, se essa é normal. No caso do MQO, temos distribuição assintótica normal dos estimadores sob as seguintes hipóteses:
* Observações $\left(y_{i}, \mathbf{x}_{i}\right), i=1, \ldots, n$, i.i.d.
* $\mathbb{E}\left[y^{2}\right]<\infty$
* $\mathbb{E}\left[\|x\|^{2}\right]<\infty$
* $\mathbf{Q}_{\mathrm{xx}}=\mathrm{E}[\mathbf{x x}]$ positiva definida.
* $\mathbb{E}\left[y^{4}\right]<\infty$ e $\mathbb{E}\left[\left\|x_{i}\right\|^{4}\right]<\infty$.

Sob A1-A5, conforme, $n \rightarrow \infty, \sqrt{n}(\widehat{\beta}-\beta) \stackrel{d}{\rightarrow} N\left(0, V_{\beta}\right)$, onde $V_{\beta}=\mathrm{Q}_{\mathrm{xx}}^{-1} \Omega Q_{\mathrm{xx}}^{-1}, \mathrm{Q}_{\mathrm{xx}}=\mathbb{E}\left[\mathrm{x}_{i} x_{i}\right]$ e $\Omega=\mathbb{E}\left[x_{i} x_{i} e_{i}^{2}\right]$. Todas as amostras são i.i.d, assim a hipótese A1 é satisfeita para todas. Dada as distribuições dos erros dos três modelos, sabemos que temos momentos finitos para os modelos 1 e 2, enquanto que para a distribuição de Cauchy isso não é garantido, fazendo com que não possamos inferir sobre consistência para os estimadores de OLS do modelo 3. A hipótese 4 vale para os três: $\mathbf{x}=\left[\begin{array}{ll}1 & x\end{array}\right]$. Assim $\mathbf{x x}^{\prime}=\left[\begin{array}{l}1 \\ x\end{array}\right] \cdot\left[\begin{array}{ll}1 & x\end{array}\right]=\left[\begin{array}{ll}1 & x \\ x & x^{2}\end{array}\right]$ e portanto
$$
\mathbb{E}\left[\mathbf{x x}^{\prime}\right]=\left[\begin{array}{rr}
	1 & \mathbb{E}(x) \\
	\mathbb{E}(x) & \mathbb{E}\left(x^{2}\right)
\end{array}\right]=\left[\begin{array}{rr}
	1 & 5 \\
	5 & 26
\end{array}\right]<\infty
$$
E vejamos se a hipótese 5 vale para esses dois modelos:

\begin{equation*}
\begin{split}
	\mathbb{E}\left(\|\mathbf{x}\|^{4}\right)=\left[\begin{array}{r}1\\ \mu^{4}+6 \mu^{2} \sigma^{2}+3 \sigma^{4}\end{array}\right]=\left[\begin{array}{r}1 \\ 2650\end{array}\right]<\infty
\end{split}
\end{equation*}
pois $x \sim(5,1)$ e $E\left[e^{(1)^{4}}\right]<\infty$ e $E\left[e^{(2)^{4}}\right]<\infty$, pois o primeiro erro é uma mistura de distribuições com quarto momento finito e o segundo porque a distribuição gama com os parâmetros definidos aqui tem quarto momento finito menos $\beta_{0}$. Quaisquer $a$ e $b$ reais, sabemos que $(a+b)^{4}=a^{4}+4 a^{3} b+6 a^{2} b^{2}+4 a b^{3}+b^{4}$. Tome $a=x$ e $b=\beta_{0}+e^{(i)}$ e note que isso implica que $E[a+b]<\infty$, pois $a$ e $b$ são independentes de modo que a esperança da multiplicação equivale a multiplicação das esperanças, além disso, quarto momento finito implica finitude de momentos menores. Assim, a hipótese 5 vale para os modelos 1 e 2 e a distribuiçāo assintótica é como definida acima.

Suponha que você conheça as distribuições verdadeiras. No modelo 1, assuma que p = 1/2. No modelo 3, assuma que as inovações tenham distribuição t-Student, com o número de graus de liberdade desconhecido (a ser estimado). Obtenha o estimador de MV para a amostra original do exercício 2.

Utilizando as funções de log-likelihood através do exercício 1 para os modelos 1 e 2, excluindo o termo $\sum_{i=1}^{n} f\left(x_{i}\right)$, dado que ele some ao tomarmos a derivada no parâmetro $\theta^{(i)}$ e não gera nenhuma perda.

Relembrando alguns parâmetros definidos no começo:

* $\beta_{0} = 1.5$
* $\beta_{1} = 3$
* $\mu_{1} = -2$
* $\mu_{2} = 2$
* $p = 0.5$
* $\sigma_{1}^2 = 0.5$
* $\sigma_{2}^2 = 25$

E a função log-likelihood, ocultando $\sum_{i=1}^{n} f\left(x_{i}\right)$:

\begin{equation*}
\begin{split}
\begin{aligned}
	\log L\left(\theta^{(1)}\right)=& \sum_{i=1}^{n} \log \left\{\frac{1}{2} \frac{1}{\sigma_{1} \sqrt{2 \pi}} \exp \left(\frac{-\left(y_{i}^{(1)}-\left(\beta_{1} x_{i}+\beta_{0}+\mu_{1}\right)\right)^{2}}{2 \sigma_{1}^{2}}\right)\right\}+\\ & \sum_{i=1}^{n} \log \left\{\frac{1}{2} \frac{1}{\sigma_{2} \sqrt{2 \pi}} \exp \left(\frac{-\left(y_{i}^{(1)}-\left(\beta_{1} x_{i}+\beta_{0}+\mu_{2}\right)\right)^{2}}{2 \sigma_{2}^{2}}\right)\right\}
\end{aligned}
\end{split}
\end{equation*}

Por motivos de identificação, setaremos dois outros parâmetros

* $\mu_3 = \beta_0 + \mu_1 = 1.5 + (-2) = -0.5$
* $\mu_4 = \beta_0 + \mu_2 = 1.5 + (2) = 3.5$

```python
# função MLE para o Modelo 1
def mle_normal(parameters, x, y, p):
   beta_1, mu_3, mu_4, sigma2_1, sigma2_2 = parameters # parametros (são os outputs futuros do problema de otimização)
   # vamos iterar a pdf de cada amostra
   n = len(x)
   LL = 0
   for i in range(n):
      pdf_i = (p * (1 / (2 * math.pi * sigma2_1) ** 0.5) * math.exp(-(y[i] - beta_1 * x[i] - mu_3) ** 2 / (2 * sigma2_1))) + \
         ((1-p) * (1 / (2 * math.pi * sigma2_2) ** 0.5) * math.exp(-(y[i] - beta_1 * x[i] - mu_4) ** 2 / (2 * sigma2_2)))
      if pdf_i == 0:
         return np.inf
      LL = LL + math.log(pdf_i)
   # Calcula o negativo da log-likelihood
   neg_LL = -LL
   return neg_LL
```

```python
# função de otimização para o modelo 1
def estimadores_mle_normal(x, y, chute_inicial):
    """
    Calcula os valores que minimiza a função de verossimilhança dada a amostra
    :param x: valores de x da amostra de tamanho n
    :param y: valores de y da amostra de tamanho n
    :param chute_inicial: chute inicial, na forma (beta_1, mu_3, mu_4, sigma2_1, sigma2_2)
    :return: tupla com os valores ótimos (beta_1, mu_3, mu_4, sigma2_1, sigma2_2)
    """
    estimadores_mle = minimize(
        mle_normal,
        x0 = chute_inicial, 
        args = (x, y, p),
        bounds = ((None, None), (None, None), (None, None), (0.01, 10), (0.01, None)),
        ).x
    return estimadores_mle
```

E a função log-likelihood para uma distribuição Gamma:

\begin{equation*}
\log L\left(\theta^{(2)}\right)=-n \log \left(\Gamma(a) b^{a}\right)+(a-1) \sum_{i=1}^{n} \log \left(y_{i}^{(2)}-\beta_{0}-\beta_{1} x_{i}\right)-\frac{1}{b} \sum_{i=1}^{n}\left(y_{i}^{(2)}-\beta_{0}-\beta_{1} x_{i}\right)
\end{equation*}

```python
# função de otimização para o modelo 2
def verossimilhanca_inovacoes_gama(parametros_gamma, x, y):
    beta_0, beta_1, a, b = parametros_gamma
    n = len(x)
    LL = 0
    for i in range(n):
        e = y[i] - beta_0 - beta_1*x[i]
        if e <= 0:
            return np.inf
    LL = LL - math.log(math.gamma(a) * b ** a) + (a-1)*math.log(e) - (e)/b
    return -LL
```

```python
# função de otimização para o modelo 2
def estimadores_inovacoes_gama(x, y, chute_inicial):
    """
    Calcula os valores que minimiza a função de verossimilhança dada a amostra
    :param x: valores de x da amostra de tamanho n
    :param y: valores de y da amostra de tamanho n
    :param chute_inicial: chute inicial na forma (beta_0, beta_1, a, b)
    :return: tupla com os valores ´otimos (beta_0, beta_1, a, b)
    """
    while True:
        x0 =[np.random.normal(chute_inicial[i], 0.2) for i in range(4)]
        otimos = minimize(
            verossimilhanca_inovacoes_gama,
            x0=x0,
            args=(x, y),
            bounds=((None, None), (None, None), (0.00001, None), (0.00001, None)),
            method="L-BFGS-B"
            )
        if otimos.success and (otimos.x != x0).all():
            break
    return otimos.x
```

Relembrando a log-likelihood da distribuição Cauchy, ocultando $\sum_{i=1}^{n} f\left(x_{i}\right)$:
\begin{equation*}
\log L\left(\theta^{(3)} \mid r=1\right)=\sum_{i=1}^{n}\left\{\log \frac{1}{\pi\left(1+\left(y_{i}^{(3)}-\beta_{0}-\beta_{1} x_{i}\right)^{2}\right)}\right\}
\end{equation*}

```python
# função de otimização para o modelo 3
def verossimilhanca_inovacoes_cauchy(parametros, x, y):
    beta_0, beta_1 = parametros
    n = len(x)
    LL = 0
    for i in range(n):
        LL = LL + math.log(1 / (math.pi * (1 + (y[i] - beta_0 - beta_1 * x[i]) ** 2)))
    return -LL
```

```python
# função de otimização para o modelo 3
def estimadores_inovacoes_cauchy(x, y, chute_inicial):
    """
    Calcula os valores que minimiza a função de verossimilhançaa dada a amostra
    :param x: valores de x da amostra de tamanho n
    :param y: valores de y da amostra de tamanho n
    :param chute_inicial: chute inicial na forma (beta_0, beta_1)
    :return: tupla com os valores ótimos (beta_0, beta_1)
    """
    estimadores_mle = minimize(verossimilhanca_inovacoes_cauchy,
        x0=chute_inicial,
        args=(x, y),
        bounds=((None, None), (None, None)),
        ).x
    return estimadores_mle
```

### 6.2 Root Mean Squared Error (RMSE)

O RMSE é uma medida para avaliar o ajuste dos valores preditivos em relação aos verdadeiros. A fórmula da RMSE é dada por

\begin{equation*}
    \operatorname{RMSE} = \sqrt{\frac{\sum_{i=1}^n \left(\hat{y_i}-y_i\right)^2}{n}}
\end{equation*}


Para Facilitar os cálculos, vamos definir algumas variáveis de ambiente

##### OLS

```python
# modelo 1
m1_ols_beta0 = beta1_bootstrap[:,0]
m1_ols_beta1 = beta1_bootstrap[:,1]
```

```python
# modelo 2
m2_ols_beta0 = beta2_bootstrap[:,0]
m2_ols_beta1 = beta2_bootstrap[:,1]
```

```python
# modelo 3
m3_ols_beta0 = beta3_bootstrap[:,0]
m3_ols_beta1 = beta3_bootstrap[:,1]
```

##### MLE

Para a computação do valor de RMSE do estimador de $\beta_0$ do modelo 1, vamos utilizar o fato de que $\mu_3 = \beta_0 + \mu_1$ e $\mu_4 = \beta_0 + \mu_2$. Usando o fato que $\mu_1 + \mu_2 = (-2) + 2 = 0$, podemos calcular

\begin{equation*}
    \frac{\mu_3 + \mu_4}{2} = \frac{(\beta_0 + \mu_1) + (\beta_0 + \mu_2)}{2} = \beta_0
\end{equation*}

Portanto, usaremos $\frac{\hat{\mu_3} + \hat{\mu_4}}{2}$ como estimador de $\hat{\beta_0}$.

```python
# modelo 1
m1_mle_mu3 = beta1_bootstrap_mle[:,1]
m1_mle_mu4 = beta1_bootstrap_mle[:,2]
m1_mle_beta0 = (m1_mle_mu3 + m1_mle_mu4)/2
m1_mle_beta1 = beta1_bootstrap_mle[:,0]
```

```python
# modelo 2
m2_mle_beta0 = beta2_bootstrap_mle[:,0]
m2_mle_beta1 = beta2_bootstrap_mle[:,1]
```

```python
# modelo 3
m3_mle_beta0 = beta3_bootstrap_mle[:,0]
m3_mle_beta1 = beta3_bootstrap_mle[:,1]
```

Lembrando os valores dos parâmetros verdadeiros

* $\beta_0 = 1.5$
* $\beta_1 = 3$

```python
# função que calcula RMSE
def RMSE(estimador, parametro):
    n = len(estimador)
    return ((np.sum((estimador - parametro) ** 2)) / n) ** (0.5)
```

```python
# Tabela com os RMSE para os estimadores de OLS
rmse_ols = pd.DataFrame([[RMSE(m1_ols_beta0, b0), RMSE(m2_ols_beta0, b0), RMSE(m3_ols_beta0, b0)],
                         [RMSE(m1_ols_beta1, b1), RMSE(m2_ols_beta1, b1), RMSE(m3_ols_beta1, b1)]],
                        columns=['Modelo 1', 'Modelo 2', 'Modelo 3'], 
                        index=['β0', 'β1'])
rmse_ols
```

```python
# Tabela com os RMSE para os estimadores de MLE
rmse_mle = pd.DataFrame([[RMSE(m1_mle_beta0, b0), RMSE(m2_mle_beta0, b0), RMSE(m3_mle_beta0, b0)],
                         [RMSE(m1_mle_beta1, b1), RMSE(m2_mle_beta1, b1), RMSE(m3_mle_beta1, b1)]],
                        columns=['Modelo 1', 'Modelo 2', 'Modelo 3'], 
                        index=['β0', 'β1'])
rmse_mle
```

Podemos notar que os RMSE dos estimadores de MLE são menores do que os estimadores de OLS. Isso era de se esperar dado que os estimadores de Máximo Verossimilhança são estimados a partir da distribuição verdadeira (e correta) dos dados. Quando se tem a informação correta da distribuição geradora de dados o estimador de MLE é o melhor. Enquanto por outro lado, a estimação por OLS supõe normalidade dos dados, o que não é verdade sempre (nesse exercício, por exemplo). De fato, o estimador de Mínimos Quadrados Ordinários é o melhor estimador linear não viesado, mas isso quando não sabemos a distribuição verdadeira dos dados (caso mais comum no mundo real).
